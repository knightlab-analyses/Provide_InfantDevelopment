```python
from os import listdir, mkdir
import pandas as pd
import numpy as np
```

```python
def printResults(folder_path):
    filename = [f for f in listdir(folder_path) if f.endswith('.tsv')]
    d = pd.DataFrame()
    for f in filename:
        df = pd.read_csv(folder_path + '/' + f, sep='\t')
        df = df.drop(df.columns[0], axis=1)
        d = d.append(df)
        d = d[['column_name', 'effect_size', 'pval']]
        d[['effect_size', 'pval']] = d[['effect_size', 'pval']].round(5) 
    return d 
```

```python
names = ['results_allfaith','results_allshannon',
        'beta_results_allweighted', 'beta_results_allunweighted','beta_results_allaitchison',
        'beta_results_alljaccard','beta_results_allbray']
```

```python
result = pd.DataFrame({'column_name':[]})
for i in names:
    df = printResults(i)
    metric = i.split('_')[-1]
    df.columns=['column_name', metric+"_"+"eff-size", metric+"_"+"pval"]
    re = re.merge(df, how='outer')
```
